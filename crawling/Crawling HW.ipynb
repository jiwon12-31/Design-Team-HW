```python
def akbo_crawler():
    df = DataFrame()
    
    for i in range(1,101):
        ajax_url = 'http://www.akbobada.com/ajaxPartgList.html?pageIndex={}&order1=&order2=&order3=&catCode=02&groupCode=12&searchText='
        r_ajax = requests.get(ajax_url.format(i))
        r_ajax.text[1:-1]
        json_result = r_ajax.json()
        df = df.append(pd.DataFrame(json_result))
        df1 = df.iloc[:,0:5]
        df1.drop('ALBUM_ID', axis=1, inplace=True)
        df2 = df.iloc[:,0].tolist()
        
    df3 = np.empty([3000,5], dtype=object)
    
    for i in range(len(df2)):
        url = 'http://www.akbobada.com/musicDetail.html?musicID={}&p=1'
        R = requests.get(url.format(df2[i]))
        Soup = bs(R.content, 'lxml')
    
        for n in range(4,100):
            info = [value.text.strip() for value in Soup.findAll('tr')[n].findAll('td')]
            
            try : 
                info[0] == '피아노 3단 악보'
            except IndexError : pass
            else:
                if info[0] == '피아노 3단 악보':
                    break
                elif info[0] == '피아노&듀엣 악보':
                    break
                elif info[0] == 'Solo&피아노 악보':
                    break
                else :
                    pass
        
        df3[i,0] = df2[i]
        df3[i,1] = info[0]
        df3[i,2] = info[1]
        df3[i,3] = info[2]
        df3[i,4] = info[3]
    
    result_df = pd.DataFrame(df3)
    Result = pd.concat([df1, result_df], axis=1,  join_axes=[df1.index])
    Result.columns = ['id1', '곡명', '앨범', '아티스트', 'id2', '파트', '설명', 'key(조)', '페이지']
    Result.drop(['id1', 'id2'], axis=1, inplace=True)
    Result.to_csv('akbo.csv', index=False)
```

```python
akbo_crawler()
```

```python
start = time.time()
akbo_crawler()
end = time.time()
print('{} seconds'.format(end-start))
```
